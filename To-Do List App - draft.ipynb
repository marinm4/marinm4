```python
class Task:
    def __init__(self, title, description, due_date):
        self.title = title
        self.description = description
        self.due_date = due_date
        self.completed = False
        
class ToDoList:
    def __init__(self):
        self.tasks = []
        
    def add_task(self, task):
        self.tasks.append(task)
        
    def remove_task(self, task):
        self.tasks.remove(task)
        
    def mark_task_complete(self, task):
        task.completed = True
        
    def display_all_tasks(self):
        if not self.tasks:
            print('\nThere are no tasks currently in the to-do list.')
        else:
            print('To-Do List: ')
            for i, task in enumerate(self.tasks, 1):
                status = 'Completed' if task.completed else 'Not Completed'
                print(f'{i}. {task.title} ({status}) - {task.description} Due Date: {task.due_date}')
                

def main():
    todo_list = ToDoList()
    
    while True:
        print('\n==== To-Do List App ====')
        print('1. Add Task')
        print('2. Remove Task')
        print('3. Mark Task as Complete')
        print('4. Display all Tasks')
        print('5. Exit')
        
        choice = input('Enter your choice (1-5): ')
        
        if choice == '1':
            title = input('Enter name of task: ')
            description = input('Enter description of task: ')
            due_date = input('What is the due date of the task?: ')
            task = Task(title, description, due_date)
            todo_list.add_task(task)
            print('Task has been added successfully to your list.')
            
        elif choice == "2":
            if todo_list.tasks:
                print("Select a task to remove:")
                for i, task in enumerate(todo_list.tasks, 1):
                    print(f"{i}. {task.title}")
                task_number = int(input("Enter task number: "))
                if 1 <= task_number <= len(todo_list.tasks):
                    task = todo_list.tasks[task_number - 1]
                    todo_list.remove_task(task)
                    print("Task removed successfully!")
                else:
                    print("Invalid task number.")
            else:
                print("\nThere are no tasks currently in the to-do list.")

        elif choice == "3":
            if todo_list.tasks:
                print("Select a task to mark as completed:")
                for i, task in enumerate(todo_list.tasks, 1):
                    print(f"{i}. {task.title}")
                task_number = int(input("Enter task number: "))
                if 1 <= task_number <= len(todo_list.tasks):
                    task = todo_list.tasks[task_number - 1]
                    todo_list.mark_task_as_completed(task)
                    print("Task marked as completed!")
                else:
                    print("Invalid task number.")
            else:
                print("\nThere are no tasks currently in the to-do list.")

        elif choice == "4":
            todo_list.display_all_tasks()

        elif choice == "5":
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

if __name__ == "__main__":
    main()
```
